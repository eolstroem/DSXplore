## Steps 5.) and 6.) Use Python to create a report of errors

```python
import pandas as pd
import numpy as np
import datacompy
import difflib


# Read in the two excel files and fill NA
df1 = pd.read_excel('Task 4.xlsx','State System').fillna(0)
df2 = pd.read_excel('Task 4.xlsx','SIS').fillna(0)
```

```python
#Take a Quick peak of the State System dataframe, aka, df1
df1.head()
```

```python
#Take a Quick peak of the SIS dateframe, aka, df2
df2.head()
```

```python
# Change FirstName and LastName in df2 to uppercase
df2['FirstName']=df2['FirstName'].str.upper()
df2['LastName']=df2['LastName'].str.upper()
df2.head()
```

```python
# To match SIS data to State data we convert the string type for StateID from float to integer
df2['StateID']=df2['StateID'].astype(np.int64)
print(df2['StateID'])
```

# Error Report: Differences between the Two DataFrames

```python
# Use datacompy library to create a report that compares the two dataframes
compare = datacompy.Compare(
    df1,
    df2,
    join_columns='StateID',  #You can also specify a list of columns eg ['policyID','statecode']
    abs_tol=0, #Optional, defaults to 0
    rel_tol=0, #Optional, defaults to 0
)
print(compare.report())
```
